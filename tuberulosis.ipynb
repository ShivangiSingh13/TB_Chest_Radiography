```python
# Step 1: Import libraries
import cv2 as cv
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.utils import class_weight
```

```python
# Step 2: Load and preprocess data
normaldir = r'Normal'
tbdir = r'Tuberculosis'
images = []
labels = []
imagesize = 256
```

```python
for x in os.listdir(normaldir):
    imagedir = os.path.join(normaldir, x)
    image = cv.imread(imagedir, cv.IMREAD_GRAYSCALE)
    if image is not None:  
        image = cv.resize(image, (imagesize, imagesize))
        image = cv.cvtColor(image, cv.COLOR_GRAY2RGB)
        images.append(image)
        labels.append(0)
    
for y in os.listdir(tbdir):
    imagedir = os.path.join(tbdir, y)
    image = cv.imread(imagedir, cv.IMREAD_GRAYSCALE)
    if image is not None:  
        image = cv.resize(image, (imagesize, imagesize))
        image = cv.cvtColor(image, cv.COLOR_GRAY2RGB)
        images.append(image)
        labels.append(1)

images = np.array(images)
labels = np.array(labels)
```

```python
# Step 3: Split and normalize data
imagetrain, imagetest, labeltrain, labeltest = train_test_split(images, labels, test_size=0.3, random_state=42)
imagetrain = (imagetrain.astype('float32'))/255
imagetest = (imagetest.astype('float32'))/255
```

```python
# Step 4: Create validation set from training data
X_train, X_val, y_train, y_val = train_test_split(
    imagetrain, labeltrain, test_size=0.2, random_state=42, stratify=labeltrain
)

print("Train shape:", X_train.shape, y_train.shape)
print("Val shape:", X_val.shape, y_val.shape)
print("Test shape:", imagetest.shape, labeltest.shape)

# Step 5: Compute class weights (to handle class imbalance)
class_weights_array = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weights = dict(enumerate(class_weights_array))
print("Class weights:", class_weights)
```

```python
# Step 4: Apply SMOTE oversampling
'''imagetrain_flattened = imagetrain.reshape(imagetrain.shape[0], -1)
smote = SMOTE(random_state=42)
imagetrain_flattened, labeltrain = smote.fit_resample(imagetrain_flattened, labeltrain)
imagetrain = imagetrain_flattened.reshape(-1, imagesize, imagesize, 3)'''
```

```python
print("Training images shape:", imagetrain.shape)
print("Training labels shape:", labeltrain.shape)
```

```python
# Step 5: Build ResNet50 model with transfer learning
'''base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(imagesize, imagesize, 3))
base_model.trainable = False  # Freeze the base model

resnet_model = keras.Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

resnet_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)'''

# Step 7: Build ResNet50 model with transfer learning

base_model = ResNet50(
    weights='imagenet',
    include_top=False,
    input_shape=(imagesize, imagesize, 3)
)

# First freeze all
base_model.trainable = True

# Improvement 3: Unfreeze more layers (fine-tuning last ~50 layers)
for layer in base_model.layers[:100]:   # you can try 80, 100, 120 etc.
    layer.trainable = False

# Improvement 5: richer classification head
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(1, activation='sigmoid')(x)

resnet_model = keras.Model(inputs=base_model.input, outputs=predictions)

resnet_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-5),  # Improvement 4: lower LR
    loss='binary_crossentropy',
    metrics=['accuracy']
)

resnet_model.summary()
```

```python
# Step 6: Data augmentation
datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True
)

datagen.fit(X_train)
```

```python
# Step 6: Train model
'''reduce_lr = ReduceLROnPlateau(monitor='val_accuracy', factor=0.1, patience=1, min_lr=0.00001, verbose=1)

base_model.trainable = True
for layer in base_model.layers[:143]:
    layer.trainable = False
    
resnet_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-5),  # Lower learning rate
    loss='binary_crossentropy',
    metrics=['accuracy']
)


history_fine = resnet_model.fit(
    imagetrain,
    labeltrain,
    epochs=10,
    batch_size=16,
    validation_split=0.2,
    callbacks=[reduce_lr],
    verbose=2
)''' 
# Step 8: Train model with augmentation and class weights

reduce_lr = ReduceLROnPlateau(
    monitor='val_accuracy',
    factor=0.1,
    patience=2,
    min_lr=1e-6,
    verbose=1
)

epochs = 20   # Improvement 4: train longer
batch_size = 16

history_fine = resnet_model.fit(
    datagen.flow(X_train, y_train, batch_size=batch_size),
    epochs=epochs,
    validation_data=(X_val, y_val),
    class_weight=class_weights,
    callbacks=[reduce_lr],
    verbose=2
)
```

```python
# Step 9: Final evaluation on TEST set (unseen data)
predictions = resnet_model.predict(imagetest, batch_size=32)
predicted_labels = (predictions > 0.5).astype('int32')

print(classification_report(labeltest, predicted_labels, target_names=["Normal", "TB"]))

accuracy = accuracy_score(labeltest, predicted_labels)
print(f"\nTest Accuracy: {accuracy * 100:.2f}%")
```

```python
# Step 8: Plot training and validation history
plt.figure(figsize=(12, 4))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(history_fine.history['accuracy'], 'b--', label='Accuracy')
plt.plot(history_fine.history['val_accuracy'], 'r--', label='Val Accuracy')
plt.title('Accuracy Over Epochs')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.ylim([0, 1])
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history_fine.history['loss'], 'b--', label='Loss')
plt.plot(history_fine.history['val_loss'], 'r--', label='Val Loss')
plt.title('Loss Over Epochs')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Step 9: Plot confusion matrix
cm = confusion_matrix(labeltest, predicted_labels)
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Normal', 'TB'], 
            yticklabels=['Normal', 'TB'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
